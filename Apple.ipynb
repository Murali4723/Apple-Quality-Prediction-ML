```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df["Quality"] = label_encoder.fit_transform(df["Quality"])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop(["A_id", "Quality"],axis=1),df['Quality'],test_size=0.3,random_state=0)
X_train.shape,X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X_train)
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled=pd.DataFrame(X_train_scaled,columns=X_train.columns)
X_test_scaled=pd.DataFrame(X_test_scaled,columns=X_test.columns)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
log_reg = LogisticRegression()
log_reg.fit(X_train_scaled, y_train)
y_pred = log_reg.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print("Classification Report:\n", classification_rep)
```

```python
from sklearn.metrics import confusion_matrix

conf_matrix = confusion_matrix(y_test, y_pred)

conf_matrix_df = pd.DataFrame(conf_matrix,
                              index=["Actual Bad", "Actual Good"],
                              columns=["Predicted Bad", "Predicted Good"])

print(conf_matrix_df)
```

###Model Evaluation
The model was evaluated on the test set using the following metrics:

The overall accuracy of the model was 72%, meaning the model correctly classified apple quality in a significant portion of test cases.

Precision: Indicates the proportion of correct predictions for each class.

Recall: Measures the ability of the model to capture all relevant instances.

F1-Score: Harmonic mean of precision and recall, providing a balance between the two.

4.3 Confusion Matrix
The confusion matrix helps visualize true positives, false positives, true negatives, and false negatives in the predictions.

            Predicted 0   Predicted 1
    Actual 0     TP         FP
    Actual 1     FN         TN

Where:

TP (True Positive): Correctly predicted good quality apples

FP (False Positive): Incorrectly classified bad apples as good

FN (False Negative): Incorrectly classified good apples as bad

TN (True Negative): Correctly predicted bad quality apples
